```python
columns = {'자치구별(1)' : '지역',
           '2020' : '2020년 출산율',
           '2021' : '2021년 출산율',
           '2022' : '2022년 출산율'}
birth = birth.rename(columns = columns)
birth.head()
```

```python
for column in ['2020년 출산율', '2021년 출산율', '2022년 출산율']:
    birth[column] = pd.to_numeric(birth[column], errors='coerce')

birth.info()
```

```python
rate_2021 = birth[['지역', '2021년 출산율']]
rate_2022 = birth[['지역', '2022년 출산율']]
```

```python
columns = {'DL_YMD' : '보낸날짜',
           'REC_CTGG_NM' : '지역',
           'E_C_01' : '가구/인테리어',
           'E_C_02' : '기타',
           'E_C_03' : '도서/음반',
           'E_C_04' : '디지털/음반',
           'E_C_05' : '생활/건강',
           'E_C_06' : '스포츠/레저',
           'E_C_07' : '식품',
           'E_C_08' : '출산/육아',
           'E_C_09' : '패션의류',
           'E_C_10' : '패션잡화',
           'E_C_11' : '화장품/미용'
            }
cj_2022 = cj_2022.rename(columns = columns)
cj_2021 = cj_2021.rename(columns = columns)
```

```python
item2021 = cj_2021.groupby('지역')['출산/육아'].sum().reset_index()
item2021
```

```python
item2022 = cj_2022.groupby('지역')['출산/육아'].sum().reset_index()
item2022
```

```python
birth_2021 = item2021.merge(rate_2021, on = '지역')
birth_2021
```

```python
birth_2022 = item2022.merge(rate_2022, on='지역')
birth_2022
```

```python
birth_re_2021 = birth_2021.rename(columns = {'출산/육아' : '2021년 출산/육아 택배량'})
birth_re_2022 = birth_2022.rename(columns = {'출산/육아' : '2022년 출산/육아 택배량'})
```

```python
people_2021 = pd.read_csv(DATA_PATH + '/서울시 2021년 구별 인구.csv')
people_2022 = pd.read_csv(DATA_PATH + '/서울시 2022년 구별 인구.csv')
```

```python
birth_people_2021 = birth_re_2021.merge(people_2021, on = '지역')
birth_people_2021
```

```python
birth_people_2022 = birth_re_2022.merge(people_2022, on = '지역')
birth_people_2022
```

```python
birth_people_2021['2021년 1인당 출산/육아 택배량'] = birth_people_2021['2021년 출산/육아 택배량'] / birth_people_2021['합계']
birth_people_2021.head()
```

```python
birth_people_2021 = birth_people_2021.drop(['합계'], axis = 1)
```

```python
birth_people_2021.head()
```

```python
birth_people_2022['2022년 1인당 출산/육아 택배량'] = birth_people_2022['2022년 출산/육아 택배량'] / birth_people_2022['합계']
birth_people_2022.head()
```

```python
birth_people_2022 = birth_people_2022.drop(['합계'], axis = 1)
birth_people_2022.head()
```

```python
import statsmodels.api as sm

X = birth_people_merge['2021년 출산율']
y = birth_people_merge['2021년 1인당 출산/육아 택배량']

# 상수항(절편)을 추가
X = sm.add_constant(X)

# 회귀 모델을 생성
model = sm.OLS(y, X).fit()

# 회귀 모델 요약 정보 출력
print(model.summary())
```

```python
X1 = birth_people_merge['2022년 출산율']
y1 = birth_people_merge['2022년 1인당 출산/육아 택배량']

# 상수항(절편)을 추가
X1 = sm.add_constant(X1)

# 회귀 모델을 생성
model1 = sm.OLS(y1, X1).fit()

# 회귀 모델 요약 정보 출력
print(model1.summary())
```
