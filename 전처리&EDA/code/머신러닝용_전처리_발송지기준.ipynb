# 추가전처리

## 월/일/요일 추가

```python
send_data['DATE'] = pd.to_datetime(send_data['DATE'])
send_data['MONTH'] = send_data['DATE'].dt.month
send_data['DAY'] = send_data['DATE'].dt.day
```

```python
send_data.head()
```

```python
columns_to_move = ['MONTH', 'DAY']
desired_positions = [2, 3]

for col, pos in zip(columns_to_move, desired_positions):
    column = send_data.pop(col)
    send_data.insert(pos, column.name, column)
send_data.head()
```

```python
send_data['DAY_OF'] = send_data['DATE'].dt.day_name()
send_data.head()
```

```python
column_to_move = send_data.columns[19]
desired_position = 4

column = send_data.pop(column_to_move)
send_data.insert(desired_position, column.name, column)
send_data.head()
```
