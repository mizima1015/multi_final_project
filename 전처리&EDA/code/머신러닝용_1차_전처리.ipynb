# 배송지/수신지 기준 전처리 전 전처리

```python
cj_2021 = cj_2021.drop(['SEND_CTPV_CD', 'SEND_CTGG_CD', 'REC_CTPV_CD', 'REC_CTGG_CD'], axis =1)
cj_2022 = cj_2022.drop(['SEND_CTPV_CD', 'SEND_CTGG_CD', 'REC_CTPV_CD', 'REC_CTGG_CD'], axis =1)
cj_2023 = cj_2023.drop(['SEND_CTPV_CD', 'SEND_CTGG_CD', 'REC_CTPV_CD', 'REC_CTGG_CD'], axis =1)
```

```python
cj_2021.head()
```

```python
cj_2021['DL_YMD'] = pd.to_datetime(cj_2021['DL_YMD'], format='%Y%m%d')
cj_2022['DL_YMD'] = pd.to_datetime(cj_2022['DL_YMD'], format='%Y%m%d')
cj_2023['DL_YMD'] = pd.to_datetime(cj_2023['DL_YMD'], format='%Y%m%d')
```

# 배송지 서울 기준 전처리

## 배송지 서울인 데이터 고르기

```python
send_2021 = cj_2021[cj_2021['SEND_CTPV_NM'] == '서울특별시']
send_2021.head()
```

```python
count_desired_data = len(send_2021[(send_2021['REC_CTPV_NM'] == '서울특별시') & (send_2021['REC_CTGG_NM'].isnull())])
print(count_desired_data)
```

```python
send_2022 = cj_2022[cj_2022['SEND_CTPV_NM'] == '서울특별시']
send_2022.head()
```

```python
count_desired_data = len(send_2022[(send_2022['REC_CTPV_NM'] == '서울특별시') & (send_2022['REC_CTGG_NM'].isnull())])
print(count_desired_data)
```

```python
send_2023 = cj_2023[cj_2023['SEND_CTPV_NM'] == '서울특별시']
send_2023.head()
```

```python
count_desired_data = len(send_2023[(send_2023['REC_CTPV_NM'] == '서울특별시') & (send_2023['REC_CTGG_NM'].isnull())])
print(count_desired_data)
```

## 열 이름 바꾸기

```python
columns = {'DL_YMD' : 'DATE',
           'SEND_CTPV_NM' : 'SEND_CITY',
           'SEND_CTGG_NM' : 'SEND_GU',
           'REC_CTPV_NM' : 'REC_CITY',
           'REC_CTGG_NM' : 'REC_GU',
           'E_C_01' : 'FURNITURE/INTERIOR',
           'E_C_02' : 'ETC',
           'E_C_03' : 'BOOK/ALBUM',
           'E_C_04' : 'DIGITAL/HOME APP',
           'E_C_05' : 'LIFE/HEALTH',
           'E_C_06' : 'SPORTS/LEISURE',
           'E_C_07' : 'FOOD',
           'E_C_08' : 'CHILDBIRTH/PARENTING',
           'E_C_09' : 'FASHION/CLOTHES',
           'E_C_10' :	'FASHION/ACC',
           'E_C_11' : 'COSMETICS/BEAUTY'}
send_2021 = send_2021.rename(columns = columns)
send_2022 = send_2022.rename(columns = columns)
send_2023 = send_2023.rename(columns = columns)
```

```python
send_2021.head()
```

## 보기 편하게 수신지(구) 위치변경

```python
column_to_move = send_2021.columns[15]
desired_position = 4

# 열을 추출하고 새로운 위치로 삽입
column = send_2021.pop(column_to_move)
send_2021.insert(desired_position, column.name, column)
send_2021.head()
```

```python
column_to_move = send_2022.columns[15]
desired_position = 4


column = send_2022.pop(column_to_move)
send_2022.insert(desired_position, column.name, column)
send_2022.head()
```

```python
column_to_move = send_2023.columns[15]
desired_position = 4

column = send_2023.pop(column_to_move)
send_2023.insert(desired_position, column.name, column)
send_2023.head()
```

## 추후 추가할 변수 데이터를 위해 연도 추가

```python
send_2021['YEAR'] = send_2021['DATE'].dt.year
send_2022['YEAR'] = send_2022['DATE'].dt.year
send_2023['YEAR'] = send_2023['DATE'].dt.year
```

```python
column_to_move = send_2021.columns[16]
desired_position = 1

column = send_2021.pop(column_to_move)
send_2021.insert(desired_position, column.name, column)
send_2021.head()
```

```python
column_to_move = send_2022.columns[16]
desired_position = 1

column = send_2022.pop(column_to_move)
send_2022.insert(desired_position, column.name, column)
send_2022.head()
```

```python
column_to_move = send_2023.columns[16]
desired_position = 1

column = send_2023.pop(column_to_move)
send_2023.insert(desired_position, column.name, column)
send_2023.head()
```

## 합치기

```python
data = [send_2021, send_2022, send_2023]
send_data = pd.concat(data)
send_data
```

```python
send_data = send_data.reset_index(drop=True)
send_data
```
