```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import numpy as np
```

```python
data_2020 = pd.read_csv('/content/2020년출산율.csv')
data_2021 = pd.read_csv('/content/2021년출산율.csv')
data_2022 = pd.read_csv('/content/2022년출산율.csv')
```

```python
data_2020.head()
```

```python
data_2020 = data_2020.set_index('REC_GU')['2020년 출산율']  # REC_GU를 인덱스로 설정하고 출산율을 데이터로 사용
data_2021 = data_2021.set_index('REC_GU')['2021년 출산율']
data_2022 = data_2022.set_index('REC_GU')['2022년 출산율']
```

```python
data_2020
```

```python
data_2021
```

```python
data_2022
```

```python
combined_data = pd.concat([data_2020, data_2021, data_2022], axis=1)

# 컬럼 이름 변경
combined_data.columns = ['2020', '2021', '2022']

# 결과 확인
print(combined_data.head())
```

```python
combined_data
```

```python
# 각 지역별로 예측된 데이터를 저장할 딕셔너리 생성
forecast_2023 = {}

for gu in combined_data.index:  # 각 지역별로 반복
    # 데이터 가져오기
    time_series_data = combined_data.loc[gu]  # 시계열로 활용할 데이터

    # ARIMA 모델 훈련
    model = ARIMA(time_series_data, order=(5, 1, 0))  # 파라미터는 실제 데이터에 따라 조절되어야 합니다.
    model_fit = model.fit()

    # 2023년 예측
    forecast_2023[gu] = model_fit.forecast(steps=1)[0]  # steps=1: 1년치(2023년) 예측

# 결과 출력
print(forecast_2023)
```

```python
data = {
    '구': forecast_2023.keys(),
    '2023년_출산율': forecast_2023.values()
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 데이터프레임 확인
print(df)
```

```python
df
```

```python
df = df.rename(columns  = {'구' : 'REC_GU'})
df.head()
```
