# 회귀분석

```python
import statsmodels.api as sm

X = s_2021['2021년 출산율']
y = s_2021['2021 1인당 출산/육아 택배량']

# 상수항(절편)을 추가
X = sm.add_constant(X)

# 회귀 모델을 생성
model = sm.OLS(y, X).fit()

# 회귀 모델 요약 정보 출력
print(model.summary())
```

```python
X1 = s_2022['2022년 출산율']
y1 = s_2022['2022 1인당 출산/육아 택배량']

# 상수항(절편)을 추가
X1 = sm.add_constant(X1)

# 회귀 모델을 생성
model = sm.OLS(y1, X1).fit()

# 회귀 모델 요약 정보 출력
print(model.summary())
```

# 피어슨 상관계수, t-검정

```python
from scipy.stats import pearsonr, ttest_ind
```

```python
vari_2021_1 = s_2021['2021년 출산율']
vari_2021_2 = s_2021['2021 1인당 출산/육아 택배량']

correlation, p_value = pearsonr(vari_2021_1, vari_2021_2)
print("피어슨 상관계수:", correlation)

t_stat, t_p_value = ttest_ind(vari_2021_1, vari_2021_2)
print("t-검정 p-value:", t_p_value)
if p_value < 0.05:
    print("두 변수 간의 관계는 통계적으로 유의미합니다.")
else:
    print("두 변수 간의 관계는 통계적으로 유의미하지 않습니다.")
```

```python
vari_2022_1 = s_2022['2022년 출산율']
vari_2022_2 = s_2022['2022 1인당 출산/육아 택배량']

correlation1, p_value1 = pearsonr(vari_2022_1, vari_2022_2)
print("피어슨 상관계수:", correlation1)

t_stat1, t_p_value1 = ttest_ind(vari_2022_1, vari_2022_2)
print("t-검정 p-value:", t_p_value1)
if p_value1 < 0.05:
    print("두 변수 간의 관계는 통계적으로 유의미합니다.")
else:
    print("두 변수 간의 관계는 통계적으로 유의미하지 않습니다.")
```
