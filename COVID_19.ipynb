### 구간 구분
- 거리두기 시행 중과 종료를 종료일 기준으로 나눠 컬럼으로 구분한다.

```python
digi_21['거리두기']='O'
digi_23['거리두기']='X'
digi_22['거리두기'] = None

```

```python
digi_22_before = digi_22[digi_22['DL_YMD']<20220418].copy()
digi_22_after = digi_22[digi_22['DL_YMD']>=20220418].copy()
digi_22_before['거리두기'] = 'O'
digi_22_after['거리두기'] =  'X'
digi_22_before.dropna(axis=1)
digi_22_after.dropna(axis=1)
```

```python
before=pd.concat([digi_21, digi_22_before])
after=pd.concat([digi_22_after, digi_23])
```

### 시각화

```python
import numpy as np
quantity_bf=before['E_C_04'].sum()
quantity_af=after['E_C_04'].sum()

plt.figure(figsize=(10,6))
plt.bar(['사회적 거리두기 기간 가전 물량', '사회적 거리두기 이후 가전 물량'], [quantity_bf, quantity_af])
plt.title(['사회적 거리두기 기간/이후 (2022년 4월 18일 이전 / 이후) 디지털/가전 물량비교'])

current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])

x = np.arange(2)
year = ['사회적 거리두기 이전', '사회적 거리두기 이후'] # x축 표시
value = [quantity_bf, quantity_af]

for i, v in enumerate(x):
    plt.text(v, value[i], value[i],
             fontsize = 11,
             fontweight = 'bold',
             color = 'royalblue',
             horizontalalignment='center',
             verticalalignment='bottom')
plt.ylabel('물량')
plt.show()
```
