```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/concatenated_df.csv')
df.info()
```

```python
print(df)
```

```python
df['DATE'] = df['DATE'].astype(str)
df['DATE'] = df['DATE'].apply(lambda x: f"{x[:4]}-{x[4:6]}-{x[6:]}")
```

```python
df['DATE']
```

```python
# 'DATE' 컬럼을 Datetime 형식으로 변환
df['DATE'] = pd.to_datetime(df['DATE'], format='%Y-%m-%d')

df.head()
```

```python
# 'YEAR' 컬럼을 추가
df['YEAR'] = df['DATE'].dt.year

df.head()
```

## 연도별 요일별 물류량 계산

```python
df['DAY_NAME'] = df['DATE'].dt.day_name()

day_name_mapping = {
    'Monday': '월',
    'Tuesday': '화',
    'Wednesday': '수',
    'Thursday': '목',
    'Friday': '금',
    'Saturday': '토',
    'Sunday': '일'
}

# '요일' 열을 영어에서 한국어로 매핑
df['DAY_NAME'] = df['DAY_NAME'].map(day_name_mapping)
df.head()
```

```python
# 연도별 요일별 평균 물류량 계산
average_by_year_day = df.groupby(['YEAR', 'DAY_NAME'])['TOTAL'].mean().reset_index()
average_by_year_day
```

```python
average_by_year_day = average_by_year_day.rename(columns={'TOTAL':'DAY_NAME_AVERAGE'})
average_by_year_day
```

```python
# 연도별 요일별 평균을 원본 데이터프레임에 추가
df = pd.merge(df, average_by_year_day, on=['YEAR', 'DAY_NAME'], how='left')

df.info()
```

```python
df = df.rename(columns={'DITIGAL/HOME APPLIANCE':'DIGITAL/HOME APPLIANCE'})
```

## 품목별 요일별 물류량 평균 구하기

```python
prod = ['FURNITURE/INTERIOR', 'ETC', 'BOOK/RECORD', 'DIGITAL/HOME APPLIANCE',
                 'LIFE/HEALTH', 'SPORTS', 'FOOD', 'MATERNITY SUPPLIES', 'CLOTHING',
                 'MISCELLANEOUS GOODS', 'COSTMETICS/BEAUTY']
```

```python
r = df.groupby(['YEAR', 'DAY_NAME']).mean()
result = r[prod].reset_index()
result.head()
```
