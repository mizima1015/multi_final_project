```python
import matplotlib.pyplot as plt

# 판매량 데이터 준비
periods = ['5월 12일 - 5월 25일', '4월 28일 - 5월 11일']
sales = [may_late_period_sum, april_to_may_early_period_sum]

# 막대 그래프를 그릴 준비
fig, ax = plt.subplots(figsize=(8, 6))

# 막대 그래프 생성
ax.bar(periods, sales, color=['blue', 'orange'])

# 타이틀 및 라벨 설정
ax.set_title('기간별 디지털/가전 물류량 비교')
ax.set_xlabel('기간')
ax.set_ylabel('물류량')

# y축의 범위를 150000부터 210000까지로 설정하여 확대 효과를 줍니다.
ax.set_ylim(150000, 210000)

# 그래프에 값 표시
for i, v in enumerate(sales):
    # 값이 확대된 범위 안에 있는 경우에만 표시합니다.
    if 150000 <= v <= 210000:
        ax.text(i, v + 500, str(v), color='black', ha='center')

# 그래프를 화면에 표시
plt.tight_layout()
plt.show()
```
